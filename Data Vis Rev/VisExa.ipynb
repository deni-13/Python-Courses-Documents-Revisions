```python
import matplotlib .pyplot as plt
import numpy as np

%matplotlib inline

x=np.linspace(0,10,100)

plt.plot(x,np.sin(x),"b")
plt.plot(x,np.cos(x),"m") 

plt.xlabel("x")
plt.ylabel("y")
plt.legend(("sin","cos"))
```

```python
plt.subplot(2,1,1) #satir,sutun,panel  #--->hepsÅŸni alt alta gostermek


plt.plot(x,np.sin(x),"b",linewidth=4)
plt.subplot(2,1,2) #satir,sutun,panel  #--->hepsini alt alta gostermek

plt.plot(x,np.cos(x),"m") 

```

```python
 #line plot


import seaborn as sns
import pandas as pd
```

```python
ir=sns.load_dataset("iris")
```

```python
sns.lineplot(data=ir,y="petal_length",x=ir.index,hue="species") #hue turlerin ayÄ±rdÄ±

#data,y eks,x eks-->index,hue
```

```python
 #scatter plot--> korealasyon
```

```python
fig,ax=plt.subplots()

ax.scatter(ir["sepal_length"],ir["sepal_width"])
#turlere gore
ax.set_title("iris dataset")

ax.set_xlabel("sepal_length")
ax.set_ylabel("sepal_width")
```

```python
#box plot

ir.groupby(by="species").mean()["sepal_length"] #turlere gore ort
```

```python
data=ir.groupby(by="species").mean()["sepal_length"] 
ax=sns.barplot(x=data.index,y=data.values)  #ort 'a gore cizdirdik sepal lentghleri
ax.set_ylabel("mean")
```
